```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("assets/shopmania_dataset.csv")
df.head(10)
```

```python
X_text = df["Product_Name"]
```

```python
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(X_text)
```

```python
y = df["Category"]
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X, y)
```

```python
df["predicted_category"] = model.predict(X)
```

```python
df.to_csv("tagged_products.csv", index=False)
print(df[["Product_Name", "predicted_category"]])
```

```python
new_df = pd.read_csv("assets/input_new_products.csv")
new_df.head(10)
```

```python
X_new = vectorizer.transform(new_df["Product Name"])
```

```python
new_df["predicted_category"] = model.predict(X_new)
```

```python
new_df.to_csv("output_new_products_tagged.csv", index=False)
print(new_df)
```
